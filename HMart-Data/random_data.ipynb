```python
import json
import random
from bson import ObjectId
from datetime import datetime, timedelta
```

```python
with open('data/users.json', 'r') as file:
    users = json.load(file)

with open('data/products.json', 'r', encoding='utf-8') as file:
    products = json.load(file)

with open('data/addresses.json', 'r', encoding='utf-8') as file:
    addresses = json.load(file)
```

```python
def random_address():
    address = {}

    province = addresses[0]['province']
    address['province'] = province

    district = random.choice(addresses[0]['districts'])
    address['district'] = district['name']

    ward = random.choice(district['wards'])
    address['ward'] = ward['name']

    address['street'] = str(random.randint(1,999))
    
    return address
```

```python
def random_date():
    start_date = datetime(2020, 1, 1)
    random_seconds = random.randint(0, int((datetime.now() - start_date).total_seconds()))
    random_date = start_date + timedelta(seconds=random_seconds)
    return random_date.strftime("%Y-%m-%dT%H:%M:%S.%fZ")[:-4] + "Z"
```

```python
def random_data_user_group(id_brand, head, last, name_brand):
    list_user = []
    for user in users[head:last]:
        userjson = {
            "_id": {"$oid": user['_id']['$oid']},
            "firstname": user['firstname'],
            "lastname": user['lastname'],
            "mobile": user['mobile']
        }
        list_user.append(userjson)

    brand_product = []
    other_product = []
    for product in products:
        info_product = {
            '_id': {"$oid": product['_id']['$oid']},
            'colors': product['colors'],
            'price': product['price']
        }
        if (product['brand']['$oid'] == id_brand):
            brand_product.append(info_product)
        else:
            other_product.append(info_product)

    ratings = []
    wishlists = []
    orders = []
    carts = []

    for user in list_user:
        address = random_address()

        wishlist = {
            "user": {"$oid": user['_id']['$oid']},
            'wishlist': []
        }

        cart = {
            "user": {"$oid": user['_id']['$oid']},
            'carts': []
        }

        the_rest_product = brand_product[:]

        num_ratings = random.randint(20, 25)
        for _ in range(num_ratings):
            product = random.choice(the_rest_product)
            the_rest_product.remove(product)
            
            star = random.randint(3, 5)
            comment = create_comment(True)
            rating = {
                "user": {"$oid": user['_id']['$oid']},
                "product": product['_id'],
                "star": star,
                "comment": comment
            }
            ratings.append(rating)
            
            orderAt = random_date()
            typePayment = random.choice(['COD', 'VNPAYQR', 'VNBANK', 'INTCARD'])
            order = {
                "user": {"$oid": user['_id']['$oid']},
                'orderItems': [],
                'shippingInfo': {
                    'firstname': user['firstname'],
                    'lastname': user['lastname'],
                    'mobile': user['mobile'],
                    'province': address['province'],
                    'district': address['district'],
                    'ward': address['ward'],
                    'street': address['street'],
                },
                'totalPrice': 0,
                'orderAt': {"$date": orderAt},
                'status': 'Delivered',
                'typePayment': typePayment
            }

            color = random.choice(product['colors'])
            quantity = random.randint(1,2)
            product_order = {
                "product": product['_id'],
                "color": color,
                "price": product['price'],
                "quantity": quantity
            }
            totalPrice = quantity * product['price']
            order['orderItems'].append(product_order)
            order['totalPrice'] = totalPrice
            orders.append(order)

        num_ratings = random.randint(5, 10)
        num_executions = random.randint(3, 4)
        executed_count = 0
        for _ in range(num_ratings):
            product = random.choice(other_product)
            
            star = random.randint(1, 2)
            comment = create_comment(False)
            rating = {
                "user": {"$oid": user['_id']['$oid']},
                "product": product['_id'],
                "star": star,
                "comment": comment
            }
            ratings.append(rating)

            orderAt = random_date()
            typePayment = random.choice(['COD', 'VNPAYQR', 'VNBANK', 'INTCARD'])
            order = {
                "user": {"$oid": user['_id']['$oid']},
                'orderItems': [],
                'shippingInfo': {
                    'firstname': user['firstname'],
                    'lastname': user['lastname'],
                    'mobile': user['mobile'],
                    'province': address['province'],
                    'district': address['district'],
                    'ward': address['ward'],
                    'street': address['street'],
                },
                'totalPrice': 0,
                'orderAt': {"$date": orderAt},
                'status': 'Delivered',
                'typePayment': typePayment
            }

            color = random.choice(product['colors'])
            quantity = random.randint(1,2)
            product_order = {
                "product": product['_id'],
                "color": color,
                "price": product['price'],
                "quantity": quantity
            }
            totalPrice = quantity * product['price']
            order['orderItems'].append(product_order)
            order['totalPrice'] = totalPrice
            orders.append(order)

            if executed_count < num_executions:
                wishlist['wishlist'].append({"product": product['_id']})
                executed_count += 1


        num_ratings = random.randint(10, 15)
        for _ in range(num_ratings):
            product = random.choice(the_rest_product)

            wishlist['wishlist'].append({"product": product['_id']})

            object_id = str(ObjectId())
            color = random.choice(product['colors'])
            quantity = random.randint(1,2)
            product_cart = {
                "_id": {"$oid": object_id},
                "product": product['_id'],
                "color": color,
                "price": product['price'],
                "quantity": quantity
            }
            cart['carts'].append(product_cart)

        wishlists.append(wishlist)
        carts.append(cart)
        
    formatted_json = json.dumps(orders, indent=4, sort_keys=False, ensure_ascii=False)
    with open('json/' + name_brand + '/orders.json', 'w', encoding='utf-8') as file:
        file.write(formatted_json)

    formatted_json = json.dumps(ratings, indent=4, sort_keys=False, ensure_ascii=False)
    with open('json/' + name_brand + '/ratings.json', 'w', encoding='utf-8') as file:
        file.write(formatted_json)

    formatted_json = json.dumps(wishlists, indent=4, sort_keys=False, ensure_ascii=False)
    with open('json/' + name_brand + '/wishlists.json', 'w', encoding='utf-8') as file:
        file.write(formatted_json)

    formatted_json = json.dumps(carts, indent=4, sort_keys=False, ensure_ascii=False)
    with open('json/' + name_brand + '/carts.json', 'w', encoding='utf-8') as file:
        file.write(formatted_json)
```
